# Извлечение именованных сущностей из текста на русском языке с использованием предобученной модели BERT

В этом блокноте демонстрируется, как с помощью предобученной модели BERT извлекать именованные сущности (NER) из текста на русском языке. Для этого используется инструмент `pipeline` из библиотеки Hugging Face Transformers.

## Введение

Предобученные модели на основе BERT позволяют решать задачи распознавания именованных сущностей. В данном примере мы используем модель, предобученную на русском языке (например, `cointegrated/rubert-tiny`), для извлечения сущностей из текстов на русском.

```python
from transformers import pipeline

# Загружаем pipeline для задачи NER
ner_pipeline = pipeline("ner", model="zaalbar/rubert-finetuned-ner", tokenizer="zaalbar/rubert-finetuned-ner")
print('Pipeline для NER успешно загружен.')
```

## Пример текста

Рассмотрим следующий пример текста на русском языке:

```python
text = "Вчера доктор Иванов выписал рецепт на Аспирин, который помогает при головной боли."
print(text)
```

## Извлечение сущностей

Запустим pipeline для извлечения именованных сущностей из текста.

```python
results = ner_pipeline(text)
for entity in results:
    print(entity)
```

## Объяснение результатов

Каждый элемент в списке `results` — это словарь, содержащий информацию о найденной сущности:

- **word**: выделенный текст сущности
- **score**: вероятность принадлежности к данной сущности
- **entity**: тип сущности (например, `B-DRUG`, `I-DOCTOR` и т.д.)
- **index**: позиция токена в тексте

Модель самостоятельно определяет, где начинается и где продолжается сущность.
