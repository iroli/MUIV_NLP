```python
import torch
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline

# Загрузка предобученной модели и токенизатора для NER на русском языке
model_name = "xlm-roberta-large-finetuned-conll03-english"  # Поддерживает много языков, включая русский
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)

# Создание пайплайна для NER
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple")

# Текст на русском языке для анализа
text = "Илон Маск является генеральным директором компании Tesla и основателем SpaceX."

# Выполнение NER
ner_results = ner_pipeline(text)

# Вывод результатов
for entity in ner_results:
    print(f"Сущность: {entity['word']}, Метка: {entity['entity_group']}, Начало: {entity['start']}, Конец: {entity['end']}")
```
